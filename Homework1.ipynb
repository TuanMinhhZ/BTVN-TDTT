Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.


```python
a, b, c = map(int, input().split())
if a+b>c and a+c>b and b+c>a:
  p=(a+b+c)/2
  s= (p*(p-a)*(p-b)*(p-c))**0.5
  print(s)
else:
  print("Khong phai 3 canh cua tam giac")
```

Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng

```python
ten_chu_ho = input("Ten chu ho: ")
chi_so_truoc = int(input("Chi so thang truoc: "))
chi_so_nay = int(input("Chi so thang nay: "))
so_dien = chi_so_nay - chi_so_truoc
tien_dien = 0
if so_dien > 400:
    tien_dien += (so_dien - 400) * 3460
    so_dien = 400
if so_dien > 300:
    tien_dien += (so_dien - 300) * 3350
    so_dien = 300
if so_dien > 200:
    tien_dien += (so_dien - 200) * 2998
    so_dien = 200
if so_dien > 100:
    tien_dien += (so_dien - 100) * 2380
    so_dien = 100
if so_dien > 50:
    tien_dien += (so_dien - 50) * 2050
    so_dien = 50
if so_dien > 0:
    tien_dien += so_dien * 1984
tong_tien = tien_dien * 1.08
tien_phai_tra = int(round(tong_tien))
print(f"Ho va ten: {ten_chu_ho}")
print(f"Tien phai tra la: {tien_phai_tra}")
```

9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
n=int(input())
match n:
  case 0:
    print("Không")
  case 1:
    print("Một")
  case 2:
    print("Hai")
  case 3:
    print("Ba")
  case 4:
    print("Bốn")
  case 5:
    print("Năm")
  case 6:
    print("Sáu")
  case 7:
    print("Bảy")
  case 8:
    print("Tám")
  case 9:
    print("Chín")
  case _:
    print("Nhập từ 1->9 thôi!")
```
