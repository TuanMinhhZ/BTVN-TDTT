Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->
int:


```python

def Max_Of_Two(a , b)->int:
  if a>b:
    return a
  else:
    return b
a=15
b=35
c = Max_Of_Two(a, b)
print(c)
```

Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]:

```python
def swap(a, b) -> Tuple[int, int]:
  a, b = b, a
  return a, b
x = 5
y = 10
x, y = swap(x, y)
print(f"x = {x}, y = {y}")
```
